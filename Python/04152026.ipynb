### Example - Is Prime

```python
def isprime(n):
    if n <= 1:
        print("Not a prime")
        return
    
    for i in range(2, n):
        if n % i == 0:
            print("Not a prime")
            return
    
    print("Prime")
```

```python
isprime(4)
```

### Generators
Return end the function and sends back the value but yield pauses the fucntion and remebers the state , allowing it to resume later

```python
def gen(n):
    for i in range(n):
        return i**3
```

```python
gen(10)
```

```python
def gen(n):
    for i in range(n):
        yield i**3
```

```python
gen(10)
```

```python
for i in gen(9):
    print(i)
```

### Filter

```python
# Returns only True value
num = [5, 12, 17, 18, 24, 32]

def t(x):
  if x < 18:
    return False
  else:
    return True

p = filter(t, num)

for x in p:
  print(x)
```

```python
list(map(lambda x:x%2==0,num))
```

```python
list(filter(lambda x:x%2==0,num))
```
